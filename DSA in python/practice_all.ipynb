```python
class linked_list():
    def __init__(self):
        self.head = None
        self.tail=None
        self.n=0

    def add_front(self,data):
        if not self.head:
            self.head = Node(data)
            self.tail = self.head
            self.n=self.n+1
        else:
            new_node = Node(data)
            self.head.prev=new_node
            new_node.next = self.head
            self.head = new_node
            self.n=self.n+1

    def append(self,data):
        if not self.head:
            self.head = Node(data)
            self.tail = self.head
            self.n=self.n+1
        else:
            
            self.tail.next = Node(data)
            self.tail.next.prev = self.tail
            self.tail = self.tail.next
            self.n=self.n+1
    def add_by_value(self,data,value):
        if value==self.head.data:
            self.add_front(data)
        # elif index==self.tail.n:
        #     self.append(data)
        else:
            new_node = Node(data)
            current = self.head
            while current:
                if current.data == value :

                    break
                else:
                    current = current.next

            if current:

                new_node.next = current.next
                new_node.prev=current
                current.next = new_node
                self.n=self.n+1
                return new_node.data
            else:
                print("Value not found")


    def del_front(self):
        self.head=self.head.next
        self.n=self.n-1
    
    def del_last(self):
        self.tail=self.tail.prev
        self.tail.next=None
        self.n=self.n-1
        

    def del_value(self,data):
        if self.head.data==data:
            self.del_front()
            return data
        elif self.tail.data==data:
            self.del_last()
            return data
        else:
            current = self.head
            while current:
                if current.data == data :
                    break
                else:
                    current=current.next
                        
            if current:
                current.prev.next = current.next
                current.next.prev = current.prev
                return data
            else:
                print("Value not found")
    def del_index(self,index):
        if index==0:
            self.del_front()
            return self.head.data

        else:
            i=0

            current = self.head
            while current:
                if i == index :
                    break
                else:
                    i=i+1
                    current=current.next
                        
            if current:
                current.prev.next = current.next
                current.next.prev = current.prev
                return index
            else:
                print("Value not found")

    def find_index(self,index):
        if index==0:
            self.del_front()
            return self.head.data

        else:
            i=0

            current = self.head
            while current:
                if i == index :
                    return current.data
                else:
                    i=i+1
                    current=current.next
                        

    def print(self):
        current = self.head
        while current:
            print(current.data)
            current = current.next

            

    
    def find_indextovalue(self,index):
        if index==0:
            self.del_front()
            return self.head.data

        else:
            i=0

            current = self.head
            while current:
                if i == index :
                    return current.data
                else:
                    i=i+1
                    current=current.next
                        
                    
    def find_valuetoindex(self,data):
        i=0
        current = self.head
        while current:
            if current.data == data :
                return i
            else:
                current=current.next
                i=i+1

```

```python
class stack:
    def __init__(self,limit):
        self.limit =limit-1
        self.head=None
        self.n=0

    def push(self,item):
        new_node=Node(item)
        if self.n<self.limit:
            if self.head:
                new_node.next=self.head
                self.head =new_node
                self.n += 1
            else:
                self.head = new_node
        else:
            raise ValueError("stack is full")

    def is_empty(self):
        if self.n==-1:
            return True
        else:
            return False
    def pop(self):
        if self.head:
            self.head=self.head.next
            self.n=self.n-1
        else:
            return self.is_empty()
    
    def is_full(self):
        if self.n==self.limit:
            return True
        else:
            return False
        
    def top(self):
        print(self.head.data)

    def print(self):
        temp=self.head
        while temp:
            print(temp.data,end=" ")
            temp=temp.next

    
```

```python
class queue:
    def __init__(self):
        self.head=None
        self.tail=None
        self.n=0

    def enqueue(self,data):
        new_node = Node(data)
        if self.head:
            self.tail.next=new_node
            self.tail=new_node
          
        else:
            self.head=new_node
            self.tail=new_node


    def dequeue(self):
        if self.head:
            data = self.head.data
            self.head = self.head.next
            self.n -= 1
            
            if self.head is None:  
                self.tail = None
            
            return data
        else:
            raise ValueError("empty queue")
        
    def peak(self):
        if self.head:
            return self.head.data
        else:
            raise ValueError("empty queue")
        
    def print(self):
        cur=self.head
        while cur:
            print(cur.data,end=" ")
            cur=cur.next

    
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


class bst:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert(data, self.root)

    def _insert(self, data, root):
        if data < root.data:
            if root.left is None:
                root.left = Node(data)
            else:
                self._insert(data, root.left)
        else:
            if root.right is None:
                root.right = Node(data)
            else:
                self._insert(data, root.right)

    def delete(self, value):
        self.root = self._delete(self.root, value)

    def _delete(self, root, value):
        if root is None:
            return root
        if value < root.data:
            root.left = self._delete(root.left, value)
        elif value > root.data:
            root.right = self._delete(root.right, value)
        else:
            if root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            else:
                min_node = self._min(root.right)
                root.data = min_node.data
                root.right = self._delete(root.right, min_node.data)
        return root

    def _min(self, node):
        current = node
        while current.left is not None:
            current = current.left
        return current

    def _max(self, node):
        current = node
        while current.right is not None:
            current = current.right
        return current

    def size(self):
        return self._size(self.root)

    def _size(self, node):
        if node is None:
            return 0
        return 1 + self._size(node.left) + self._size(node.right)

    def inorder(self):
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node, result):
        if node is not None:
            self._inorder(node.left, result)
            result.append(node.data)
            self._inorder(node.right, result)

    def preorder(self):
        result = []
        self._preorder(self.root, result)
        return result

    def _preorder(self, node, result):
        if node is not None:
            result.append(node.data)
            self._preorder(node.left, result)
            self._preorder(node.right, result)

    def postorder(self):
        result = []
        self._postorder(self.root, result)
        return result

    def _postorder(self, node, result):
        if node is not None:
            self._postorder(node.left, result)
            self._postorder(node.right, result)
            result.append(node.data)
```
